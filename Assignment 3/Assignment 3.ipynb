```python

# Import necessary libraries
from imutils import paths as path_util  # For working with image paths
import numpy as np  # For working with arrays
import cv2  # For image processing
import argparse  # For parsing command-line arguments
import imutils  # For image processing and resizing
import matplotlib.pyplot as plt  # For displaying images
```

```python
# Create an empty list to store images
image_collection = []
```

```python
# Load 5 images, convert them to RGB format, and store them in the list
for i in range(5):
    # Construct the path of the image file
    image_path = str(i + 1) + ".jpg"
    # Load the image
    img = cv2.imread(image_path)
    # Convert the image from BGR to RGB format
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Append the image to the image collection
    image_collection.append(img)
    # Display the image
    plt.figure()
    plt.imshow(img)
```

```python
# Create a stitcher object
stitcher_obj = cv2.Stitcher_create()
```

```python
# Stitch the images in the list
(status, stitched_image) = stitcher_obj.stitch(image_collection)
```

```python
# If the stitching was successful, display the stitched image
if status == 0:
    plt.imshow(stitched_image)
    print("Final Output")
```
